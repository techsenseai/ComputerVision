## 1. Cargar y visualizar imágenes

```python
from PIL import Image
import matplotlib.pyplot as plt

# Cargar una imagen desde archivo
image = Image.open('image.png')  # Cambia esto por la ruta de tu imagen

# Mostrar la imagen original
plt.imshow(image)
plt.title("Imagen Original")
plt.axis("off")
plt.show()
```

## 2. Convertir a escala de grises

```python
# Convertir la imagen a escala de grises
gray_image = image.convert('L')

# Mostrar la imagen en escala de grises
plt.imshow(gray_image, cmap='gray')
plt.title("Imagen en Escala de Grises")
plt.axis("off")
plt.show()
```

## 3. Convertir a un array NumPy

```python
import numpy as np

# Convertir la imagen a un array NumPy
image_array = np.array(gray_image)

# Mostrar dimensiones y valores del array
print("Dimensiones del array:", image_array.shape)
print(image_array)
```

### 3.1 Cómo funcionan las imágenes digitales
Las imágenes digitales son matrices de píxeles donde:

    - Escala de grises: Cada píxel tiene un valor entre 0 (negro) y 255 (blanco).
    - Imágenes a color: Cada píxel tiene tres valores para los canales RGB (Rojo, Verde, Azul).
    
Podemos visualizar estas matrices para entender su estructura:

```python
# Visualizar una sección de la matriz de una imagen en escala de grises
print(image_array[50:55, 50:55])  # Muestra un recorte de 5x5 píxeles
```

Para imágenes a color:

```python
# Imagen en RGB
image_color_array = np.array(image)
print(image_color_array[50:55, 50:55, :])  # Muestra los valores RGB de un recorte
```

## 4. Operaciones básicas en imágenes

```python
# Invertir colores de la imagen
inverted_image = 255 - image_array

# Mostrar la imagen invertida
plt.imshow(inverted_image, cmap='gray')
plt.title("Imagen Invertida")
plt.axis("off")
plt.show()
```

### 4.2 Espacios de color
Los espacios de color determinan cómo se representan los colores en una imagen. Algunos espacios comunes son:

    - RGB: Colores primarios (Rojo, Verde, Azul).
    - HSV: Tono (Hue), Saturación y Valor. Útil para segmentación por color.
    - Grayscale: Escala de grises.
Podemos convertir entre estos espacios con Pillow:

```python
# Convertir a HSV
hsv_image = image.convert('HSV')

# Convertir de HSV a escala de grises
gray_from_hsv = hsv_image.convert('L')
```

```python
# Mostrar la imagen invertida
plt.imshow(hsv_image, cmap='gray')
plt.title("Imagen HSV")
plt.axis("off")
plt.show()
# Mostrar la imagen invertida
plt.imshow(gray_from_hsv, cmap='gray')
plt.title("Gris desde HSV")
plt.axis("off")
plt.show()
```

### 4.3 Recortar imagen

```python
# Recortar la imagen (bounding box: left, upper, right, lower)
cropped_image = image.crop((50, 50, 200, 200))

# Mostrar la imagen recortada
plt.imshow(cropped_image)
plt.title("Imagen Recortada")
plt.axis("off")
plt.show()
```

### 4.4 Rotar imagen

```python
# Rotar la imagen 45 grados
rotated_image = image.rotate(45)

# Mostrar la imagen rotada
plt.imshow(rotated_image)
plt.title("Imagen Rotada")
plt.axis("off")
plt.show()
```

### 4.5 Cambiar tamaño de imagen

```python
# Cambiar tamaño de la imagen
resized_image = image.resize((100, 100))

# Mostrar la imagen redimensionada
plt.imshow(resized_image)
plt.title("Imagen Redimensionada")
plt.axis("off")
plt.show()
```

### 4.6 Operaciones Numéricas Avanzadas
Con NumPy, podemos realizar operaciones como:

    - Ajustar el brillo:

```python
bright_image = np.clip(image_array + 50, 0, 255)  # Aumenta el brillo
plt.imshow(bright_image, cmap='gray')
plt.show()
```

    - Aumentar el contraste:

```python
contrast_image = np.clip(1.5 * image_array - 50, 0, 255)  # Ajusta contraste
plt.imshow(contrast_image, cmap='gray')
plt.show()
```

## 5. Guardar una imagen modificada

```python
# Convertir el array invertido de nuevo a imagen
inverted_image_pil = Image.fromarray(inverted_image)

# Guardar la imagen invertida
inverted_image_pil.save('imagen_invertida.jpg')
print("Imagen guardada como 'imagen_invertida.jpg'")
```
